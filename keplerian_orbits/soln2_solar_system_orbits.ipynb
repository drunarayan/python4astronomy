```python
# Read in the Github data that will be used for the calculations.
data1 = pd.read_csv("solar_system_orbits.csv")
data1
```

```python
a = np.array(data1['Semimajor Axis (AU)'])
P = np.array(data1['Period (y)'])
e = np.array(data1['Eccentricity'])
c = a * e
data1['Semifocal Distance (AU)'] = c
b = (a**2-c**2)**0.5
data1['Semiminor Axis (AU)'] = b
ph = e*(1-a)
data1['Perihelion Distance (AU)'] = ph
ah = e*(1+a)
data1
```
